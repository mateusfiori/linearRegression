```python
from scipy.stats import pearsonr as pearson
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

```

```python
boston = load_boston()

X = boston.data
y = boston.target

coluna = X[:,0]

coef = pearson(coluna, y)
print(coef)

df = pd.DataFrame(X)

g = sns.pairplot(df)
```
